```python
# Convert all integer columns to float in one go.

# 1
# new_df = new_df.apply(lambda x: x.astype('float') if x.dtypes in ['int16','int32','int64'] else x.astype(x.dtypes) )

# 2
new_df = new_df.apply(lambda x: x.astype('float') if pd.api.types.is_integer_dtype(x) else x)

new_df.dtypes
```

```python
# Change order_id to string type.
new_df['order_id'] = new_df['order_id'].astype('string')
new_df.dtypes
```

```python
# Create a new column called net_sales = sales - discount_value.
new_df['net_sales'] = new_df['sales'] - new_df['discount_value'].astype('float')
new_df
```

```python
# Format order_date to “dd-mm-yyyy” string format.
new_df['new_order_date'] = new_df['order_date'].dt.strftime('%d-%m-%Y')
new_df
```

```python
# Check memory usage before and after converting object columns to category.
new_df.memory_usage(deep=True)
new_df.apply(lambda x: x.astype('category') if pd.api.types.is_object_dtype(x) else x)
new_df.memory_usage(deep=True)
```

```python
# Check for any rows where conversion to datetime failed (if you re-import raw text dates).
pd.to_datetime(new_df['order_date'], errors='coerce').isna()
```

```python
# Use .astype('category') on product_sub_category and compare performance.
new_df['product_sub_category'].memory_usage(deep=True)
new_df['product_sub_category'] = new_df['product_sub_category'].astype('category')
new_df['product_sub_category'].memory_usage(deep=True)
# 621547 - 183560
```

```python
# Build a new column discount_rate = discount_value / sales and convert to %

new_df['discount_rate'] = new_df['discount_value'].astype(int) / new_df['sales']
new_df['discount_rate'] = round(new_df['discount_rate'] * 100,3)
new_df['discount_rate']
```

```python
# Convert all object columns to string type using a loop.
new_df = new_df.apply(lambda x: x.astype('string') if x.dtypes == 'object' else x)
new_df.dtypes
```

```python
# Create a new column order_decade from year (e.g., 2010s, 2020s).
new_df['order_decade'] = (new_df['order_date'].dt.year // 10) * 10
new_df
```

```python
# Find all rows where converting sales to int after a float operation gives a mismatch.
new_df['sales'] = pd.to_numeric(new_df['sales'], errors='coerce')
new_df.dtypes
```
