```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("stud.csv")
```

```python
# Select the first 10 rows of the dataset.
df[:9]
```

```python
# Display only gender, math score, and reading score columns for the first 15 rows.
df.loc[:,["math score","gender","reading score"]]
```

```python
# Retrieve the last 5 rows using iloc.
df[-5:]
```

```python
# Display every alternate row from the dataset using slicing.
df[::2]
```

```python
# Get all students who scored above 85 in math.
df.loc[df["math score"] > 85]
```

```python
# Select students where gender is "female" and reading score is above 80.
df.loc[(df["gender"] == "female") | (df["reading score"] > 80)]
```

```python
# Retrieve all students who completed the test preparation course.
df.loc[df["test preparation course"] == "completed"]
```

```python
# Filter students who have "free/reduced" lunch and scored less than 60 in writing.
df.loc[(df["lunch"] == "free/reduced") & (df["writing score"] < 60) ]
```

```python
# Extract only math score and writing score columns using loc.
df.loc[:,["math score"]]
df.iloc[:,[5]]
```

```python
# Select rows 50 to 60 and columns gender, lunch, and math score together.
df.loc[50:59,["gender","lunch","math score"]]
df.iloc[50:59,[0,3,5]]
```

```python
# Find students with math score between 70 and 90 (inclusive).
df.loc[df["math score"].between(70,90,inclusive='both')]
```

```python
# Select students who belong to group A or group B and scored above 75 in reading.
df.loc[(df["race/ethnicity"].isin(["group A","group B"])) & (df["reading score"] > 75)]
```

```python
# Display only male students who didnâ€™t complete the test preparation course.
df.loc[(df["gender"] == 'male') & (df["test preparation course"] == "none")]
```

```python
# Select top 10 rows and reverse them using indexing.
df.head(10).iloc[::-1]
```

```python
# Retrieve the rows for students whose parents have a master's degree and scores in all subjects are above 80.
score_columns = ["math score", "reading score", "writing score"]
above_80 = df[score_columns].gt(80).all(axis=1)
df.loc[(df['parental level of education'] == "master's degree") & (above_80)]
```
